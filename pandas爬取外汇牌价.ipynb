```python
import pandas as pd
```

```python
url = 'http://www.boc.cn/sourcedb/whpj/index.html'
```

```python
fx=pd.read_html(url,header=0)[1]

# pd.read_html返回的是列表；列表第二个元素即是想要爬取的信息，而且是dataframe是格式
# 若没有header，后续列名处理会多些麻烦
```

## 数据规整：

```python
# 转换数据类型
fx=fx.apply(lambda x: pd.to_numeric(x,errors='ignore')) # 转换数值型
```

```python
fx['发布日期']=fx['发布日期'].apply(pd.to_datetime)    # 转换日期
```

```python
fx.ix[:,-1]=fx.ix[:,-1].apply(pd.to_timedelta)    #转换时间戳
```
